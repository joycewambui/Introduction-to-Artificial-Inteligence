```python
import random
import time

class performance:
    def _init_(self):
        self.score=0
        self.visit=0
        
    def displayscore(self):
        print("score",+ self.score)
        p=(self.score/20)*100
        print("performance",p,"%")
        return p

class state:
    def __init__(self):
        self.cleaningmethod={"a":"light","b":"light","c":"light","d":"light"}
        self.cleaningmethod["a"]=random.choice(["light","thorough"])
        self.cleaningmethod["b"]=random.choice(["light","thorough"])
        self.cleaningmethod["c"]=random.choice(["light","thorough"])
        self.cleaningmethod["d"]=random.choice(["light","thorough"])
                            

class environment:
    def __init__(self):
        self.locations = ["a", "b","c","d"]
        self.location_condition = {"a": random.randint(0, 1),
                                   "b": random.randint(0, 1),
                                   "c": random.randint(0, 1),
                                   "d": random.randint(0, 1)
                                  }

        self.vacuum_location = random.choice(self.locations)

class Agent(environment,state,performance):
    def __init__(self, environment, state,performance):
        print("Scanning location....", environment.vacuum_location)
        print("Scanning condition....", environment.location_condition)
        print("Scanning state....", state.cleaningmethod)

        count = 0
        while count < 4:
            current_location = environment.vacuum_location
            if environment.location_condition[current_location] == 1:
                if state.cleaningmethod[current_location] == "thorough":
                    state.cleaningmethod[current_location] = "light"  # Change to light after thorough cleaning
                else:
                    state.cleaningmethod[current_location] = "thorough"  # Change to thorough cleaning

                environment.location_condition[current_location] = 0  # Mark as clean
                print(current_location, "is now clean!")
                performance.score=performance.score=+1
            else:
                print("Nice!", current_location, "does not need any cleaning!")

            # Move to the next location
            newIndex = (environment.locations.index(current_location) + 1) %2

            
            environment.vacuum_location = environment.locations[newIndex]
            count += 1
            performance.visit=performance.visit=+1
        
        print("Final scanning location....", environment.vacuum_location)
        print("Final condition....", environment.location_condition)
        print("Scanning state....", state.cleaningmethod)

# Create an environment and a state
s1 = state()
p1=performance() #initialize the score to zero
x = 0
timeinterval=2
while( x <= 12):
    print("\nVisiting", x, "\n")
    e1 = environment()
    agent_instance=Agent(e1,s1,p1)
    if x==5:
       p=p1.displayscore()
       if p>50:
           print("Agent is over utilised....increasing visits and intervals")
           timeinterval=1
           x=0
       else:
           print("Agent is under utilised....decreasing visits and intervals")
           timeinterval=3
           x=x+3
        
        
    time.sleep(5)
    x = +1
    p1.displayscore()
```
